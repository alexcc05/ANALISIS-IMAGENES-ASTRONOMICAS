```python
#LINEAR REGRESSION AND PLOT
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
def func(x, a, b):
    return a*x +b

def estimate_coef(x, y,yerr, title):
    yerr = np.array([0.001,0.001,0.001])
    popt, pcov = curve_fit(func, x, y, sigma=yerr)
    perr = np.sqrt(np.diag(pcov))

    return (popt[0], perr[0],popt[1] , perr[1])
def plot_regression_line(x, y, b, title,yerr):

    plt.scatter(x, y, color = "m",
    marker = "o", s = 30)
    plt.errorbar(x, y, yerr = yerr, fmt = '.')
    y_pred = b[2] + b[0]*x

    plt.plot(x, y_pred, color = "g")

    plt.xlabel('Masa de aire', fontsize=14)
    plt.ylabel('Magnitud instrumental', fontsize=14)
    plt.gca().invert_yaxis()
    plt.title(title)
    plt.savefig(title+'.png')

    plt.show()

# observations / data
#BANDA G
y = np.array([-17.566,-17.608,-17.615])
x = np.array([1.414047259,1.16257217,1.049645528])
yerr=np.array([0.000,0.000,0.000])

# estimating coefficients
b = estimate_coef(x, y, yerr, "Banda g")
print("Estimated coefficients:\nPunto cero = {} {} \
    \nCoeficiente = {} {}".format(b[0], b[1], b[2], b[3]))

# plotting regression line
plot_regression_line(x, y, b,"Banda g",yerr)



#BANDA R
y = np.array([-17.543,-17.560,-17.558])
x = np.array([1.397220694,1.161484513,1.050006892])
yerr=np.array([0.000,0.000,0.000])

# estimating coefficients
b = estimate_coef(x, y, yerr, "Banda r")
print("Estimated coefficients:\nPunto cero = {} {} \
    \nCoeficiente = {} {}".format(b[0], b[1], b[2], b[3]))

# plotting regression line
plot_regression_line(x, y, b,"Banda r",yerr)



 #BANDA I
y = np.array([-16.851,-16.855,-16.850])
x = np.array([1.393610295,1.16033299,1.050383738])
yerr=np.array([0.001,0.000,0.000])

# estimating coefficients
b = estimate_coef(x, y, yerr, "Banda i")
print("Estimated coefficients:\nPunto cero = {} {} \
    \nCoeficiente = {} {}".format(b[0], b[1], b[2], b[3]))


# plotting regression line
plot_regression_line(x, y, b,"Banda i",yerr)



 #BANDA Z
y = np.array([-16.065,-16.068,-16.052])
x = np.array([1.389672587,1.159153177,1.050800566])
yerr=np.array([0.001,0.001,0.000])

# estimating coefficients
b = estimate_coef(x, y,yerr, "Banda z")
print("Estimated coefficients:\nPunto cero = {} {} \
    \nCoeficiente = {} {}".format(b[0], b[1], b[2], b[3]))


# plotting regression line
plot_regression_line(x, y, b,"Banda z",yerr)

```
